```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# 1. Real Estate Data (Square Feet vs Price in Lakhs)
market_data = {
    "Square_Feet": [600, 800, 1000, 1200, 1500, 1800, 2000, 2400],
    "Price_Lakhs": [30, 42, 50, 62, 78, 90, 105, 125],
}
df_house = pd.DataFrame(market_data)

# 2. Linear Regression Model Training
X = df_house[["Square_Feet"]]
y = df_house["Price_Lakhs"]
model = LinearRegression().fit(X, y)

print(
    "=================== PROJECT 3: AI HOUSING MARKET PREDICTOR ==================="
)
# Predicting for a custom square feet
test_sqft = np.array([[1650]])
pred_val = model.predict(test_sqft)
print(f"[*] Predicted Valuation for 1650 Sq.Ft House: ₹{pred_val[0]:.2f} Lakhs")
print(f"[*] Market Growth Slope Rate (Coefficient)   : {model.coef_[0]:.4f}")
print("=============================================================================\n")

# 3. Market Regression Line Plot
plt.figure(figsize=(8, 5))
plt.scatter(
    df_house["Square_Feet"],
    df_house["Price_Lakhs"],
    color="darkorange",
    label="Real Market Data",
)
plt.plot(
    df_house["Square_Feet"],
    model.predict(X),
    color="navy",
    linewidth=2,
    label="AI Regression Line",
)
plt.title("Housing Market Trend Analysis", fontsize=12, fontweight="bold")
plt.xlabel("Area in Square Feet")
plt.ylabel("Price (₹ in Lakhs)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
